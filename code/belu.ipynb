```python
from evaluate import load
bertscore = load("bertscore")
```

```python
import json
with open('./explanation_bef_inspection.json', 'r') as f:
    explanation_bef_inspection = json.load(f)

with open('./gpt_2_inter.json', 'r') as f:
    gpt_2_inter = json.load(f)
```

```python
scores = []
for i in range(len(explanation_bef_inspection)):
    i = str(i)
    try:
        label = explanation_bef_inspection[i]['label']
    except: continue
    if label == 0: continue
    gt = explanation_bef_inspection[i]['explanation']
    pred = gpt_2_inter[i]['reason']
    if str(pred) == 'None': pred = "the command is certain"
    print(i, gt, pred)
    results = bertscore.compute(predictions=[pred], references=[gt], model_type="distilbert-base-uncased")
    print(i, results)
    scores.append(results['f1'][0])
```

```python
mean = sum(scores) / len(scores)
print(mean)
```
